
## Calculating $\mu_{JT}$ for a van der Waals Gas ##


We want to write the Joule-Thomson coefficient $\mu_{JT}$ in terms of quantities that we can directly calculate from the van der Waals equation of state. Since enthalpy does *not* appear in the van der Waals equation, we'll need to get rid of it! 

To do this, let's go back to our original definition of $\mu_{JT}$ from last week: 

$$ \mu_{JT} = \left( \frac{\partial T}{\partial P} \right)_H = - \frac{\left( \frac{\partial H}{\partial P} \right)_T}{\left( \frac{\partial H}{\partial T} \right)_P} .$$

The quantity in the denominator is easy to interpret: This is simply the heat capacity
$$ C_P = \left . \frac{dQ}{dT} \right|_P = \left( \frac{\partial H}{\partial T} \right)_P $$
at constant pressure. Since this quantity is always positive and easy to measure, we'll be satisfied with this result and won't try to obtain any more explicit expression for $C_P$. 

It's less obvious what to do with the quantity $\left( \frac{\partial H}{\partial P} \right)_T$ in the numerator. To eliminate the dependence on $H$ in this term, we'll need to turn to our fundamental differential relation for enthalpy. 

### Step 0: Fundamental Relations ### 

To see how to deal with $H$ here, begin with the differential relation for energy

$$ dU = T dS - P dV $$

implied by the First and Second Laws. Note also that from the definition $H = U + PV$ for the enthalpy, we have
$$ dH = d(U + PV) = dU + P dV + V dP , $$
where in the second equality, we've used the chain rule to split apart the differential of $d(PV)$ into two terms. Plugging this into our fundamental expression for $dU$, this gives

$$ dH = T dS - P dV + P dV + V dP = T dS + V dP . $$

The is the fundamental differential relation of $H$. We'll use it to convert the partial derivative $\left( \frac{\partial H}{\partial P} \right)_T$ into terms that don't depend explicitly on enthalpy. 


### Step 1: Eliminate $H$ ###

To see how to do this, divide through the equation $dH = T dS + V dP$ by $dP$ and impose the constraint of fixed $T$ to obtain

$$ \left( \frac{\partial H}{\partial P} \right)_T = T \left ( \frac{\partial S}{\partial P} \right)_T + V .$$

Our expression for the JT coefficient is thus 

$$ \mu_{JT} = - \frac{1}{C_P} \left[  T \left ( \frac{\partial S}{\partial P} \right)_T + V \right] .$$

### Step 2: Eliminate $S$ ###

This last expression is at best a marginal improvement: We've eliminated $H$, but at the expense of introducing the entropy $S$, which isn't immediately easier to deal with. Getting rid of $S$ turns out to be straightforward, however, in light of the fundamental relation 

$$ dG = - S dT + V dP $$

for the Gibbs free energy $G$. (This relation is itself derived from our expression for $dH$ by using the definition $G = H - TS$.) As a "well-behaved" thermodynamic potential, the partial derivatives of $G$ must obey the "mixed-partial" rule or *Maxwell's Relation*
$$ \frac{\partial^2 G}{\partial P \partial T} = \frac{\partial^2 G}{\partial T \partial P} . $$
Mathematically, this means that *the order in which we take partial derivatives doesn't matter*: Taking a small step in $P$ first and then $T$ is the same as taking a small step in $T$ first and then $P$. 

In light of the fundamental relation for $dG$ above, the mixed-partials rule means that 
$$ \left ( \frac{\partial \left( -S \right) }{\partial P} \right)_T = \left ( \frac{\partial V}{\partial T} \right )_P $$
or 
$$ \left ( \frac{\partial S }{\partial P} \right)_T = - \left ( \frac{\partial V}{\partial T} \right )_P . $$

This expression looks much better: The term on the right-hand side involves only $P$, $T$, and $V$, quantities that appear already in the van der Waal's equation of state. In these terms, $\mu_{JT}$ can be expressed as 

$$ \mu_{JT} = - \frac{1}{C_P} \left[ V - T \left ( \frac{\partial V}{\partial T} \right )_P \right] .$$


### Step 3: Solve for $P$ ###

Now, we want to actually evaluate this expression using the van der Waals equation of state. We could evaluate the partial derivative $\left ( \frac{\partial V}{\partial T} \right )_P$ directly, but this is a bit messy since we would need first to solve the equation of state for $V$. An easier route is first to rewrite the expression in terms of partial derivatives of $P$, which is already solved for in our equation of state. 

To do this, we use the "cyclic permutation" rule. In general, this allows us to write partial derivatives of one variable in terms of partial derivatives of the other variables upon which it depends. In this case, since we're looking for a way to rewrite a partial derivative $\left( \frac{\partial V}{\partial T} \right)_P$ at constant $P$, we write out the formal differential for $dP$ in terms of the other two variables $V$ and $T$: 

$$ dP = \left( \frac{\partial P}{\partial T} \right)_V dT + \left( \frac{\partial P}{\partial V} \right)_T dV . $$

At constant $P$, this results in the relation

$$ \left( \frac{\partial P}{\partial T} \right)_V dT = - \left( \frac{\partial P}{\partial V} \right)_T dV \quad\quad\text{(Constant P)} . $$

Now, since we want an expression for the partial of $V$ with respect to $T$, we divide through by $dT$ to obtain

$$ \left( \frac{\partial P}{\partial T} \right)_V = - \left( \frac{\partial P}{\partial V} \right)_T \left( \frac{\partial V}{\partial T} \right)_P  $$

or 

$$ \left( \frac{\partial V}{\partial T} \right)_P = - \frac{\left( \frac{\partial P}{\partial T} \right)_V }{\left( \frac{\partial P}{\partial V} \right)_T } . $$


### Step 4: Evaluate ####
Both the numerator and denominator here are easy to evaluate from the van der Waals equation of state: 

$$ \left( \frac{\partial P}{\partial T} \right)_V  = \left( \frac{\partial \left( \frac{RT}{v - b} - \frac{a}{v^2} \right) }{\partial T} \right)_V = \frac{R}{v - b}$$

$$ \left( \frac{\partial P}{\partial V} \right)_T = \frac{1}{n} \left( \frac{\partial \left( \frac{RT}{v - b} - \frac{a}{v^2} \right)}{\partial v} \right)_T = \frac{1}{n} \left[ - \frac{RT}{\left( v - b \right)^2} + \frac{2a}{v^3} \right], $$

leaving us with an explicit -- but somewhat complicated -- expression for the JT coefficient: 
$$ \mu_{JT} = - \frac{1}{C_P} \left[ V + T \frac{\left( \frac{\partial P}{\partial T} \right)_V }{\left( \frac{\partial P}{\partial V} \right)_T } \right] = - \frac{n}{C_P} \left[ v + \frac{\frac{RT}{v - b} }{ - \frac{RT}{\left( v - b \right)^2} + \frac{2a}{v^3} } \right] .$$


### Step 5: Approximate ###

The complexity of this last expression makes it difficult to analyze. To simplify the expression, we'll make use of the fact that the coefficients $a$ and $b$ are small parameters. Note that when $a$ and $b$ are exactly zero, the Joule Thomson coefficient is also zero. We'll assume that $a$ and $b$ are small but *not quite* zero and perform a Taylor series expansion in these parameters. 

The simplest way to do this is to rearrange the denominator of the second term so that we can use the expansion 

$$ \frac{1}{1 - x} \approx 1 + x . $$

To that end, we multiply through both numerator and denominator of the second term by $- \frac{\left( v - b \right)^2}{RT}$

$$ \mu_{JT} = - \frac{n}{C_P} \left[ v + \frac{ b - v }{ 1 - \frac{\left( v - b \right)^2}{RT} \frac{2a}{v^3} } \right] \approx - \frac{n}{C_P} \left[ v + \left( b - v \right) \left( 1 + \frac{\left( v - b \right)^2}{RT} \frac{2a}{v^3} \right) \right] .$$

Finally, we expand all terms and drop any terms that contain powers of $a$ and/or $b$ higher than order one: 
$$ \mu_{JT} \approx - \frac{n}{C_P} \left[ v + b - v  - \frac{2a}{RT} \right] = - \frac{n}{C_P} \left( b - \frac{2a}{RT} \right).$$

Noting that the total heat capacity $C_P$ is related to the *specific* heat capacity $\bar C_P$ by 
$$ \bar C_P = \frac{C_P}{n},$$
this produces the explicit expression for $\mu_{JT}$ quoted in the main text.
